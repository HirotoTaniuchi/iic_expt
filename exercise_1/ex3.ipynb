```python
import numpy as np
import matplotlib.pyplot as plt


from pathlib import Path # ファイルシステム
from google.colab import drive # Google drive
drive.mount('/content/drive')
p = Path('/content/drive/My Drive/iicex/')  # 画像ファイルのフォルダ
#list( p.iterdir() )  # フォルダの内容をリストにする


from skimage import io # 入出力ライブラリの読み込み
img = io.imread('/content/drive/My Drive/iicex/ladybug.png')  # 画像ファイルの読み込み
print("img:     ", img.dtype)
print("img:     ", type(img))
print("img:     ", img.shape)
#print(img)

grayimg = np.array([[0.2989 * img[i, j, 0] + 0.5870 * img[i, j, 1] + 0.1140 * img[i, j, 2] for j in range(320)] for i in range(213)], dtype='uint8')
#320列分の1行を213行並べるので順序が逆になる
print("grayimg: ", grayimg.shape)



fig = plt.figure()
print(type(fig))
# Set original image
a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(img)
a.set_title( 'RGB')


# Set gray scaled image
a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(grayimg)
imgplot.set_cmap('gray')# 'viridis', 'inferno', 'magma', 'cividis', 'gray', 'hot'
a.set_title('Gray')
plt.colorbar(orientation='horizontal')
```
