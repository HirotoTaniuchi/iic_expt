```python
#問8

import numpy as np

# For reproducibility
rng = np.random.default_rng(0)

# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
N, D_in, H, D_out = 64, 1000, 100, 10
learning_rate = 1e-4

# Create random input data
x = rng.standard_normal((N, D_in))
# Create discrete ground truth
class_idx = np.random.randint( 0, D_out - 1, N ) # ランダムに正解クラスを生成
t = np.zeros((N,D_out)) # 他クラスは0 #64x10 #64このデータとその分類x10
t[range(N),class_idx] = 1 # 正解クラスは1

#print(x.shape, class_idx.shape, t.shape)

# Randomly initialize weights
w1 = rng.standard_normal((D_in, H))
w2 = rng.standard_normal((H, D_out))

def np_log(x): #0の対数を取らないように値を制限
    return np.log(np.clip(x, 1e-10, 1e+10))

def softmax(x):
    #x = x - np.max(x, axis=-1, keepdims = True) #オーバーフロー予防
    return np.exp(x) / np.sum(np.exp(x), axis=-1, keepdims = True) #axis = 1に注意!!行に沿って各列の和をとる(j=1~10),と考える
    #次元0(=行)に沿って総和をとったもので割るので、出力は各行の和が1になる


for i in range(500):
    # Forward pass: compute predicted y
    y1 = np.matmul(x, w1) #64x100
    y2 = np.clip(y1, 0, None) #64x100
    y_pred = np.matmul(y2, w2) #64x10
    # Calculate cross entropy loss


     
    ## Apply softmax
    y_pred_softmax = softmax(y_pred) #64x10
    #print(i, "softmax", np.sum(y_pred_softmax[0, :])) #しっかり各行の和が1.0になることの確認
    #print(i, "softmax", (y_pred_softmax[0, :]))


    ## Negative log likelihood
    #print( - np.log(y_pred_softmax ) * t)
    loss = - np.sum( t * np_log( y_pred_softmax)) #scholaor
    #各データの,正解クラスに対応する特徴量のみ(合計64ケ)を集めて合計している
    #0~1の値しか取らないのでloss>=0になるようにマイナスをつけている
    print(i, loss)


    
    # Backprop to compute gradients of w1 and w2 with respect to loss
    #grad_y_pred = y_pred_softmax - t
    grad_y_pred = y_pred_softmax
    grad_y_pred[t == 1] -= 1.0
    #print(i, "grad_y_pred", grad_y_pred[0, :])
    grad_w2 = np.matmul( np.transpose(y2), grad_y_pred)
    grad_y2 = np.matmul( grad_y_pred, np.transpose(w2))
    grad_y1 = np.copy(grad_y2)
    grad_y1[y1 <= 0] = 0.0
    grad_w1 = np.matmul(x.transpose(1,0), grad_y1)
    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

```
